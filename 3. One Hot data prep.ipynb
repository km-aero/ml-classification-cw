### Import Data

```python
f_name = dirPData + 'prepVars.pickle'

with open(f_name, "rb") as f:
    dict_ = pickle.load(f)
    
df_train = dict_['df_train']
df_test = dict_['df_test']
df_train_cat = dict_['df_train_cat']
df_train_num = dict_['df_train_num']
df_test_num = dict_['df_test_num']
del f_name, dict_
```

## Create cat test dataframe

```python
vars_test = df_test.columns.values

var_dep = ['target']

var_notToUse = ['id']
vars_ind_test = [var for var in vars_test if var not in (var_notToUse + var_dep)]
vars_ind_cat_test = list(df_test[vars_ind_test].columns[df_test[vars_ind_test].dtypes == 'object'])

df_test_cat = df_test[vars_ind_cat_test]
df_test_cat.head()
```

## Create cat test dataframe

```python
vars_train = df_train.columns.values

vars_ind_train = [var for var in vars_train if var not in (var_notToUse + var_dep)]
vars_ind_cat_train = list(df_train[vars_ind_train].columns[df_train[vars_ind_train].dtypes == 'object'])

df_train_cat = df_train[vars_ind_cat_train]
df_train_cat.head()
```

## One Hot

```python
#apply one hot
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(df_train_cat))
OH_cols_test = pd.DataFrame(OH_encoder.transform(df_test_cat))

OH_cols_train.head()
```

```python
#replace index with original one
OH_cols_train.index = df_train_cat.index
OH_cols_test.index = df_test_cat.index
```

```python
#Join one hot and numeric data
OH_df_train = pd.concat([df_train_num, OH_cols_train], axis=1)
OH_df_test = pd.concat([df_test_num, OH_cols_test],axis=1)

OH_df_train.head()
```

```python
#save variables
dict_ = {'df_test': df_test,
         'df_train': df_train,
        'df_train_num': df_train_num,
        'df_test_num': df_test_num,
        'OH_df_train': OH_df_train,
        'OH_df_test': OH_df_test}

f_name = dirPData + 'ohprepVars.pickle'
with open(f_name, "wb") as f:
    pickle.dump(dict_,f)
del f_name
```
