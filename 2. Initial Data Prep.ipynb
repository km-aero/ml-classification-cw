### Import Data

```python
f_name = dirPData + 'Vars.pickle'

with open(f_name, "rb") as f:
    dict_ = pickle.load(f)
    
df_test = dict_['df_test']
df_train = dict_['df_train']
del f_name, dict_
```

## Clean Variables
* df_train
* df_test

```python
vars_train = df_train.columns.values
vars_test = df_test.columns.values

var_dep = ['target']

var_notToUse = ['id']
vars_ind_train = [var for var in vars_train if var not in (var_notToUse + var_dep)]
vars_ind_numeric_train = list(df_train[vars_ind_train].columns[df_train[vars_ind_train].dtypes != 'object'])

df_train_num = df_train[vars_ind_numeric_train]
df_train_num.head()

vars_ind_test = [var for var in vars_test if var not in (var_notToUse + var_dep)]
vars_ind_num_test = list(df_test[vars_ind_test].columns[df_test[vars_ind_test].dtypes != 'object'])

df_test_num = df_test[vars_ind_num_test]
```

```python
vars_ind_cat_train = list(df_train[vars_ind_train].columns[df_train[vars_ind_train].dtypes == 'object'])
df_train_cat = df_train[vars_ind_cat_train]
df_train_cat.head()
```
