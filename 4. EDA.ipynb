### Check for Multicollinearity/Correlations

```python
plt.figure(figsize=(20,30))
sns.heatmap(OH_df_train.corr(), vmax=0.8)
```

The image below shows correlations of the different variables in the form of a coloured covariance matrix with **WHITE** and **PURPLE** points representing highly correlated variables.  It is somewhat difficult to see from the heatmap, but some variables definitely have correlations with others.

```python
(OH_df_train.corr()['v_num_1']).plot()
(OH_df_train.corr()['v_num_110']).plot()
(OH_df_train.corr()[104]).plot()
(OH_df_train.corr()[251]).plot()
(OH_df_train.corr()[398]).plot()
(OH_df_train.corr()[496]).plot()
plt.title("Correlation of 6 variables to all variables")
plt.legend(loc="best")
```

* Correlations between boolean variables will mostly be high
* Data includes a large amount of boolean variables even before one hot and so we shouldn't worry about high correlations between these variables.
